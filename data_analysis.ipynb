```python
import json 

file = "Data_Embedding/test.json"
test_review = None

with open(file, 'r', encoding='utf-8') as file:
        test_review = json.load(file)
    
print(len(test_review))

test_star_dict = {}
for review in test_review:
        if review["stars"] not in test_star_dict:
                test_star_dict[review["stars"]] = 1
        else: 
                test_star_dict[review["stars"]] += 1

print(test_star_dict)
```
